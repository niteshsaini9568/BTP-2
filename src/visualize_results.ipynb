```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
from image_utils import read_image, convert_rgb2gray, resize_image
```

```python
dir_test_images = "/home/abhishek/Desktop/machine_learning/coco_2017/all_test/"
dir_gen_images = "/home/abhishek/Desktop/machine_learning/coco_2017/generated_cgan_results/"
dir_save_visuals = "/home/abhishek/Desktop/machine_learning/coco_2017/visuals/"
```

```python
if not os.path.isdir(dir_save_visuals):
    os.makedirs(dir_save_visuals)
```

```python
list_test_images = sorted(os.listdir(dir_test_images))
list_gen_images = sorted(os.listdir(dir_gen_images))
```

```python
image_idx = 116
```

## Utility function to visualize/save an image

```python
def visualize_result(image_idx, save=False):
    image_test_rgb = resize_image(read_image(os.path.join(dir_test_images, list_test_images[image_idx])))
    image_test_gray = convert_rgb2gray(image_test_rgb)
    image_gen_rgb = read_image(os.path.join(dir_gen_images, list_gen_images[image_idx]))
    
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

    axes[0].imshow(image_test_gray, cmap="gray")
    axes[0].set_title("Grayscale image", fontsize=20)
    axes[1].imshow(image_test_rgb)
    axes[1].set_title("Colour image", fontsize=20)
    axes[2].imshow(image_gen_rgb)
    axes[2].set_title("Our CGAN Generated colour image", fontsize=20)
    
    if save:
        fig.savefig(os.path.join(dir_save_visuals, list_test_images[image_idx]))
        fig.clf()
    else:
        fig.show()
```

## Visualizing individual sample

```python
visualize_result(image_idx)
```

## Saving all the visualizations

```python
for i in range(0, len(list_test_images)):
    visualize_result(i, save=True)
```
